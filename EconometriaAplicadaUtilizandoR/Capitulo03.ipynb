```r
tiendas <- read.delim("cap3_datos.csv",sep=";", fileEncoding="UTF-8", header=T,stringsAsFactors=F)
```

```r
attach(tiendas)
```

```r
tiendas
```

```r
resultado <- lm(LVENTA ~ LINGR+LPR, data = tiendas)
```

```r
summary(resultado)
```

```r
residuales <- resultado$residuals
residuales
```

```r
# obtengo la matriz de varianzas-covarianzas
vcov(resultado)
```

```r
# En la tabla ANOVA podemos evaluar los
# resultados mediante el Test F
anova(resultado)
```
