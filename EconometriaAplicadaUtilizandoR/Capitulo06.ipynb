# Multicolinealidad

La multicolinealidad perfecta en realidad debe considerarse un caso poco frecuente en los modelos economÃ©tricos, que de ocurrir tendrÃ­a como consecuencia la violaciÃ³n del supuesto de rango completo de la matriz de regresores [ð—] y en consecuencia tampoco se cumplirÃ­a para la matriz [ð—â€²ð—], siendo singulares ambas matrices y sus determinantes iguales a cero, lo que darÃ­a lugar a la indeterminaciÃ³n de los estimadores de mÃ­nimos cuadrados ordinarios para los parÃ¡metros del modelo. 

El problema de la multicolinealidad debe ser visto como un problema de identificaciÃ³n, ya que alternativamente diferentes valores de los parÃ¡metros en el modelo generan el mismo valor estimado de la variable dependiente, lo que impide identificar el efecto individual de cada variable.

### SOLUCIONES AL PROBLEMA DE LA MULTICOLINEALIDAD

Si el problema de multicolinealidad no es severo, mÃ¡s vale no hacer nada, puede pasar que las acciones tomadas para remediarl la multicolinealidad, empeoren el modelo. De cualquier forma, si se quiere hacer algo para resolver el problema los remedios usuales son los siguientes: 

#### Imponer reestricciones al modelo.


Se restringen los parÃ¡metros de las variables altamente colineales.

Con x2 y x3 colineales, se detecta que el coeficiente Beta3 es un medio del coeficiente Beta2. Podemos aplicar la siguiente restricciÃ³n.
Beta3 = 0.5 * Beta2

Una vez restringido el modelo, se elimina la colinealidad. Lugo para obtener el otro parÃ¡metro, lo sustituyo en la restricciÃ³n.

La principal limitante de este mÃ©todo es la carencia de antecedentes empÃ­ricos acerca de los coeficientes de interÃ©s en los modelos economÃ©tricos.

Otra alternativa que implica restringir el modelo original es la estimaciÃ³n de un modelo en corte transversal. 
La limitante de este procedimiento es que la interpretaciÃ³n de los parÃ¡metros de corte transversal y series de tiempo puede diferir ampliamente al calcularse sobre conjuntos de datos diferentes.


#### Componentes principales

El mÃ©todo de componentes principales busca eliminar el problema de multicolinealidad a travÃ©s de la obtenciÃ³n de un conjunto de variables a partir de las originales y sin implicar grandes pÃ©rdidas de informaciÃ³n (Everitt y Hothorn, 2006). Las nuevas variables o componentes cumplen con la condiciÃ³n de ser ortogonales entre sÃ­. 

La principal limitante de este mÃ©todo es que las nuevas variables Z pueden no tener interpretaciÃ³n econÃ³mica alguna.

#### Eliminar variables

La eliminaciÃ³n de variables sospechosas de colinealidad puede ser otra opciÃ³n para evitar el problema de multicolinealidad, sin embargo puede llevarnos a un problema mÃ¡s grave como el de variable relevante omitida.

#### Transformar variables

La transformaciÃ³n de variables con primeras diferencias o calculando porcentajes es oro remedio que busca diferenciar mÃ¡s las variables entre sÃ­. Su principal limitante es que, por un lado la teorÃ­a relevante pudiera estar interesada Ãºnicamente en las variables de nivel y no en sus diferencias ni en sus porcentajes, por otro lado la variable dependiente pudiera estar relacionada con las demÃ¡s en niveles pero no en porcentajes ni en diferencias.
