## Estimación de mínimos cuadrados

```r
# Con el análisis previo se procede a estimar los parámetros de
# la ecuación 3 a través de los Mínimos cuadrados oridnarios
# MCO.
# Especifico a en R la matríz X así como el vector y.
# Se utiliza cbind()
X <- cbind(1, x2, x3)
```

```r
# Para transformar la variable deuda pública (y) a
# vector, uso el mismo código

y1 <- cbind(y) 
```

```r
# Para estimar Beta, primero obtengo
# el producto de X y su transpuesta.
# transpuesta de X
trX <- (t(X))
```

```r
# Producto de Xt y X. El producto de matrices es con %*%
X_X <- trX %*% X

X_X
```

```r
# Obtengo el determinante de la matriz X_X para determinar
# si tiene inversa o no. 
det(X_X)
```

```r
# Para obtener la inversa uso la librería MASS, 
# de la cual voy a utilizar la función ginv().
library(MASS)
```

```r
invX_X <- (ginv(X_X))
invX_X
```

```r
# Una vez que tengo la matriz inversa
# obtengo el producto de X por el vector y
Xy <-trX %*% y1
Xy
```

```r
# Calculo Beta
beta <- invX_X %*% Xy
beta
```

```r
# Para comprobar que este vector es certero, utilizo el código
# para estimar de manera directa la regresión lineal lm(y~x).
# R utiliza el mismo método
modelo <- lm(y~x2+x3)
summary(modelo)
```
