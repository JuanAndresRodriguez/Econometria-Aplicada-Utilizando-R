#  ERROR DE ESPECIFICACIÓN 

```r
# Se realiza la prueba de especificación de Ramsey , llamada
# RESET, que sirve para detectar errores de especifiación 
# ocasionados por la omisión de variables independientes,
# la posible existencia de correlación entre variables en X y E o
# porque la forma funcional de las variables independientes
# no es la apropiada

# Hay que probar
# 1) H0 forma funcional correcta
# 2) H1 forma funcional incorrecta

# Se realiza una vez que se ha estimado el modelo planteado y
# se ha calculado su ajuste.
# Consiste en agregar al modelo inicial, potencias de sus valores
# ajustados y analizar la significancia estadística conjunta
# de los parámetros asociados a las potencias de la variable
# ajustada.

# se requiere el paquete lmtest
library(lmtest)
```

```r
Gasolina <- read.csv("Gasolina.txt", header = TRUE, sep=",", encoding = "UTF-8")
```

```r
attach(Gasolina)
```

```r
cons <- lm(log(G/Pobl) ~ log(Y)+log(Pg)+log(Pan))
```

```r
resettest(cons, power=2:4)
```

```r
resettest(cons, power=2:2)
```

```r
# Los resultados de las pruebas indican que la forma funcional
# no es correcta. En ambas el p-value es menor que cualquiera
# de los niveles de significancia, comúnmente utilizados.
# El modelo debe ser reespecificado.
```

```r
# En el ejemplo dos se presenta una versión del modelo
# estático para la elasticidad de sustitución Armington para
# México. LA variable explicada en el modelo es al demanda
# relativa (DRel), que resulta del cociente entre las
# importaciones totales y la demanda doméstica (diferencia
# entre el valor bruto de la producción y las exportaciones,
# ambas a precios de mercado) y la variable explicativa es el
# precio relativo (PRel), que se obtiene de dividir el índice
# precios de la demanda reatica enre el correspondiente a las
# importaciones.
# Una vez estimado el modelo se prueba si la forma funcional
# es la correcta mediante la prueba de ramsey, RESET. Se
# realiza considerando primero la segunda y tercer potencia
# y en seguida se efectúa solo para la segunda potencia.
```

```r
Elast <- read.csv("Elast.txt")
```

```r
attach(Elast)
Elast
```

```r
model <- lm(log(DRel) ~ log(PRel))
```

```r
summary(model)
```

```r
plot(model)
```

```r
resettest(model)
```

```r
resettest(model, power=2:2)
```

```r
# En ambas pruebas no se rechaza la hipótesis nulea, el p-value
# es mayor a alfa para 1%, 5% o 10%, y por tanto hay evidencia
# a favor de que la forma funcional es correcta.
```
