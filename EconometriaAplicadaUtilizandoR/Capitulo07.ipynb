# HETEROCEDASTICIDAD

El origen  de la heterocedasticidad, está asociado a la varianza creciente de las 
perturbaciones aleatorias de  los valores de algunas de  las variables, incluidas en 
el modelo. Dicho de otro modo, podría suponerse que la varianza de la 
perturbación se compone de una parte constante, homocedastica, y otra parte 
variable según los valores de una determinada variable. Es muy probable que esta 
asociación entre el proceso de heterocedasticidad y las variables no sea evidente.  
La detección de la heterocesdasticidad en la mayoría de los procedimientos  es útil 
para establecer algún tipo de solución que permite corregir este problema. 

## Efectos de la heterocedasticidad

a) Los estimadores del MCO son lineales insesgados y consistentes, pero en precencia de heterocedasticidad son ineficientes, ya que, la varianza no es la óptima

b) Las varianzas del estimador de Mínimos Cuadrados Ordinarios, no puede calcularse con la expresión usual si se detectó heterocedasticidad. Se utiliza otra expresión.

Cuando se realizan estimaciones bajo el supuesto de que las perturbaciones 
siguen un proceso homocedástico,  mediante la aplicación de modelo de regresión 
con MCO, se cometerá un error de cálculo en la varianza, esto implica, 
básicamente, que nuestros cálculos sobre la “t” de student, ya no podrán 
comprarse con los valores de referencia correctos, y lo mismo ocurrirá con el resto 
de cálculos que tienen origen en la varianza estimada. Por ejemplo,  el contraste 
“F” ya no se distribuirá como una “F” o los contrastes que utilizan como referencia 
a la j-cuadrada. 

```r
library("readxl")
```

```r
distribucion <- read_xlsx("NumCajas.xlsx")
```

```r
attach(distribucion)
distribucion
```

```r
modeloTiempoEntrega <- lm(Tiempo ~ Cajas + Distancia, data = distribucion)
```

```r
summary(modeloTiempoEntrega)
```

```r
plot(modeloTiempoEntrega)
```

```r
# Obtengo la matriz de covarianzas
vcov(modeloTiempoEntrega)
```

```r
# Para determinar el poder explicativo del modelo utilizo el
# contraste F, al obtener la tabla ANOVA se pueden evaluar 
# los resultados mediante el test F

anova(modeloTiempoEntrega)
```

```r
# Analizo los residuales mediante análisis gráfico.
# primero obtengo el vector de residuales

residuales <- modeloTiempoEntrega$residuals
residuales
```

```r
# Reviso los supuestos de normalidad de los residuos.
rstint <- rstandard(modeloTiempoEntrega)
rstint
```

```r
# win.graph()
par(mfrow=c(1,3))
```

```r
hist(rstint)
boxplot(rstint)
```

```r
qqnorm(rstint)
```

```r
# Realizo test de Breusch-Pagan y test de White para
# detectar heterocedasticidad.

# Breusch-Pagan
library(lmtest)
bptest(modeloTiempoEntrega)
```

```r
# Uso el modelo de White. Devuelve las matrices de covarianza 
# corregidas para hacer inferencias.
library(car)
hccm(modeloTiempoEntrega)
```

```r
# Obtengo los resultados del test utilizando la matriz de
# covarianza
coeftest(modeloTiempoEntrega, vcov = hccm)
```
