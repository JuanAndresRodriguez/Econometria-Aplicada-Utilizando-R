```r
# Se crea un modelo de regresión lineal para explicar
# el Producto Bruto interno per cápita en función de su capital
PWT <- lm(PIBPCL ~ K)
```

```r
summary(PWT)
```

```r
plot(K,PIBPCL)
abline(PWT)
```

```r
# Histograma del PIB per cápita de los paises
hist(PIBPCL)
```

```r
# El histograma muestra que la mayoria de los países se 
# encuentran en los ingresos más bajos de la distribución. 
# EL histograma es útil porque muestra densidades
# (área bajo la curva igual a la unidad) y añadirle funciones
# de densidad kernel, lo cual sehace de la siguiente forma:
hist(PIBPCL,freq=FALSE)
lines(density(PIBPCL))
```

```r
# Observo la distribucion de los datos con boxplots, 
# en las cuales la caja muestra los umbrales para los 
# cuartiles inferior y superior, además de la mediana.
# Las líneas debajo y arriba permiten identificar
# observaciones extremas.
boxplot(PIBPCL)
```
